```python
X_train_mit = train.iloc[:, :-1]
y_train_mit = train.iloc[:, -1]
X_test_mit = test.iloc[:, :-1]
y_test_mit = test.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_mit_scaled = scaler.fit_transform(X_train_mit)
X_test_mit_scaled = scaler.transform(X_test_mit)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_mit_scaled, y_train_mit)
print("Label distribution before equalization (MIT-BIH Train):")
print(y_train_mit.value_counts())
print("\nLabel distribution after equalization with SMOTE:")
print(pd.Series(y_train_balanced).value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_balanced, y_train_balanced)

y_pred_mit = rf.predict(X_test_mit_scaled)

print("Accuracy on MIT-BIH Test:", accuracy_score(y_test_mit, y_pred_mit))

cm = confusion_matrix(y_test_mit, y_pred_mit)
ConfusionMatrixDisplay(cm).plot()
plt.title("Confusion Matrix for MIT-BIH")
plt.show()
```

```python
from sklearn.metrics import classification_report
print("\nClassification Report for RF:")
print(classification_report(y_test_mit, y_pred_mit))
```

```python
from sklearn.svm import SVC


svm = SVC(kernel='rbf', random_state= 42, class_weight=None)  
svm.fit(X_train_balanced, y_train_balanced)


y_pred_svm = svm.predict(X_test_mit_scaled)


print("The accuracy on MIT-BIH Test:", accuracy_score(y_test_mit, y_pred_svm))
print(classification_report(y_test_mit, y_pred_svm))


cm_svm_balanced = confusion_matrix(y_test_mit, y_pred_svm)
ConfusionMatrixDisplay(cm_svm_balanced).plot(cmap='Blues')
plt.title("Confusion Matrix for SVM ")
plt.show()
```
