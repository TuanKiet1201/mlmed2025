```python
def check_missing_values(df):
    print("\n--- Checking for Missing Values ---")
    missing_summary = df.isnull().sum()
    print("\nMissing values per column:")
    print(missing_summary)
    total_missing = missing_summary.sum()
    print(f"\nTotal missing values in the dataset: {total_missing}")

def check_duplicates(df):
    print("\n--- Checking for Duplicated Rows ---")
    duplicate_rows = df[df.duplicated()]
    print(f"Number of duplicate rows: {len(duplicate_rows)}")
    if len(duplicate_rows) > 0:
        print("\nDuplicated rows:")
        print(duplicate_rows)

if train is not None:
        check_missing_values(train)
        check_duplicates(train)
```

```python
# Lấy cột nhãn (cột cuối cùng) cho MIT-BIH
labels_train = train.iloc[:, -1]  # Cột cuối cùng
labels_test = test.iloc[:, -1]

# Phân phối nhãn trong MIT-BIH Train
print("\nPhân phối nhãn trong MIT-BIH Train:")
print(labels_train.value_counts())

# Phân phối nhãn trong MIT-BIH Test
print("\nPhân phối nhãn trong MIT-BIH Test:")
print(labels_test.value_counts())

# Đối với PTB (giả sử nhãn là cột cuối cùng, nếu không, bạn cần kiểm tra lại)
labels_ptb_normal = ptbdb_normal.iloc[:, -1] if ptbdb_normal.shape[1] > 1 else pd.Series(0)  # Giả sử nhãn 0
labels_ptb_abnormal = ptbdb_abnormal.iloc[:, -1] if ptbdb_abnormal.shape[1] > 1 else pd.Series(1)  # Giả sử nhãn 1

print("\nPhân phối nhãn trong PTB Normal:")
print(labels_ptb_normal.value_counts())
print("\nPhân phối nhãn trong PTB Abnormal:")
print(labels_ptb_abnormal.value_counts())
```
